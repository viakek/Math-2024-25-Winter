5. Check if the function $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the second-order partial differential equation (the so-called "wave equation"), where $v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda}$:

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$


Check if the function $$\psi(t,x) = A\cos(\omega t + kx)$$ is a solution of the wave equation:
$$ \frac{\partial^2 \psi(t,x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t,x)}{\partial x^2} = 0 $$
where $$ v = \frac{\omega}{k} = \frac{2\pi}{T} \frac{2\pi}{\lambda} $$.

**Compute the second partial derivatives**

Second derivative with respect to time:
$$ \frac{\partial^2 \psi}{\partial t^2} = -\omega^2 A\cos(\omega t + kx) $$

Second derivative with respect to space:
$$ \frac{\partial^2 \psi}{\partial x^2} = -k^2 A\cos(\omega t + kx) $$

**Substitute into the wave equation**

$$ -\omega^2 A\cos(\omega t + kx) - v^2 (-k^2 A\cos(\omega t + kx)) = 0 $$

$$ -\omega^2 A\cos(\omega t + kx) + v^2 k^2 A\cos(\omega t + kx) = 0 $$

Factor out $$ A\cos(\omega t + kx) $$:
$$ A\cos(\omega t + kx) (-\omega^2 + v^2 k^2) = 0 $$

Since $$ A\cos(\omega t + kx) $$ is not always zero, we must have:
$$ -\omega^2 + v^2 k^2 = 0 $$
$$ \omega^2 = v^2 k^2 $$

Since $$ v = \frac{\omega}{k} $$, the equation holds, proving that $$\psi(t,x)$$ satisfies the wave equation
