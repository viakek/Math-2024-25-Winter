1. Calculate:
   - $\displaystyle \lim_{n \to \infty} \frac{n^2 + 3n}{2 n^2 - 2n}$



$$
\frac{n^2 + 3n}{2n^2 - 2n} = \frac{\frac{n^2}{n^2} + \frac{3n}{n^2}}{\frac{2n^2}{n^2} - \frac{2n}{n^2}} = \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}.
$$


$$
\lim_{n \to \infty} \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}} = \frac{1 + 0}{2 - 0} = \frac{1}{2}.
$$


$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} = \frac{1}{2}.
$$

2. Prove using the squeeze theorem:
   - $\displaystyle\lim_{n \to \infty} \frac{\sin(n)}{n}$
To prove:

The Squeeze Theorem states:

If we have three sequences a_n, b_n, and c_n such that:
$$
a_n ≤ b_n ≤ c_n for all n ≥ N,
$$
and if:
$$
lim_{n → ∞} a_n = lim_{n → ∞} c_n = L,
$$
then:
$$
lim_{n → ∞} b_n = L.
$$


The sine function satisfies:
$$
-1 ≤ sin(n) ≤ 1, for all real n.
$$

Dividing this inequality by n > 0, we get:
$$
-1 / n ≤ sin(n) / n ≤ 1 / n.
$$



$$
lim_{n → ∞} (-1 / n) = 0.
$$


$$
lim_{n → ∞} (1 / n) = 0.
$$

both  0 as n → ∞.


Since:
$$
-1 / n ≤ sin(n) / n ≤ 1 / n,
$$
and:
$$
lim_{n → ∞} (-1 / n) = lim_{n → ∞} (1 / n) = 0,
$$

$$
lim_{n → ∞} sin(n) / n = 0.
$$
