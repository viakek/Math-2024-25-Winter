4. Find the asymptotes of the function:
  
   - $f(x) = \frac{x^2 - 1}{x^2 + 1}$
   - $g(x) = \frac{\sin(x)}{x^2+1}$

We want to find if there are any horizontal or vertical asymptotes for this function.

#### **Horizontal Asymptote of f(x)**

1. To find the **horizontal asymptote**, we look at what happens to f(x) when $$ x \to \infty $$.
   
   - Both the numerator $$ x^2 - 1 $$ and the denominator $$ x^2 + 1 $$ are dominated by the $$ x^2 $$ term when $$ x $$ is large. So, for very large values of $$ x $$we can ignore the smaller parts like -1  and  +1 .
   
   - So, for large  x , the function behaves like this:

   $$
   f(x) \approx \frac{x^2}{x^2} = 1
   $$

   - Therefore, as $$ x \to \infty $$, f(x)  gets closer and closer to 1.

   **Conclusion**: The horizontal asymptote is y = 1 .

#### **Vertical Asymptote of f(x)**

2. To find the **vertical asymptote**, we check if the denominator can become zero.

   - The denominator is $$ x^2 + 1 $$, which is never zero for any real value of x (because $$ x^2 + 1 $$ is always positive).

   **Conclusion**: There are **no vertical asymptotes** because the denominator never becomes zero.

### Conclusion for  f(x) :
- **Horizontal Asymptote**:  y = 1
- **No Vertical Asymptotes**.

 $$ g(x) = \frac{\sin(x)}{x^2 + 1}$$

Now, let's look at the function $$ g(x) = \frac{\sin(x)}{x^2 + 1} $$.

#### **Horizontal Asymptote of  g(x)**

1. To find the **horizontal asymptote**, we look at what happens as $$ x \to \infty $$.

   - The numerator is $$ \sin(x) $$, which always stays between -1 and 1 (it never goes beyond these values).
   - The denominator is $$ x^2 + 1 $$, which gets bigger and bigger as ( x ) increases.

   So, as x gets really large, the numerator stays bounded (between -1 and 1), but the denominator grows very large. This means the entire fraction gets smaller and smaller and approaches 0.

   **Conclusion**: The horizontal asymptote is  y = 0 .

#### **Vertical Asymptote of g(x)**

2. To find the **vertical asymptote**, we check if the denominator can become zero.

   - The denominator is $$ x^2 + 1 $$, which is never zero for any real value of $$ x $$ (since $$ x^2 + 1 > 0 $$ for all x ).

   **Conclusion**: There are **no vertical asymptotes** because the denominator never becomes zero.

### Conclusion for \( g(x) \):
- **Horizontal Asymptote**: y = 0
- **No Vertical Asymptotes**.
