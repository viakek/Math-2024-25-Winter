3. Let $f(x) = e^x$ and $g(x) = \ln(x)$. Check: $f(g(x))$ and $g(f(x))$. What do you notice?

1. **Finding f(g(x)):**
$$
f(g(x)) = f(ln(x)) = e^{ln(x)}.
$$
Using the property e^{ln(x)} = x, we have:
$$
f(g(x)) = x.
$$

2. **Finding g(f(x)):**
$$
g(f(x)) = g(e^x) = ln(e^x).
$$
Using the property ln(e^x) = x, we have:
$$
g(f(x)) = x.
$$

Both compositions f(g(x)) and g(f(x)) simplify to x. This means:
$$
f(g(x)) = x \quad \text{and} \quad g(f(x)) = x.
$$

### What Do We Notice?
This indicates that f(x) and g(x) are inverse functions of each other. When we compose an exponential function e^x with its natural logarithm ln(x), we get back the original input x. The same is true when we compose ln(x) with e^x. This is a key property of inverse functions:

$$
f(g(x)) = g(f(x)) = x.
$$
