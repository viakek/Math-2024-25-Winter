### Equations of Lines on a Plane

In geometry, the equation of a line on a 2D Cartesian plane describes all the points that lie on that line. The equation of a line can be represented in various forms, depending on the given information about the line.

---

### 1. **Slope-Intercept Form**  
**Equation:**  
$$
y = mx + b
$$  
- **\(m\)**: Slope of the line (rise over run, or the rate of change of \(y\) with respect to \(x\)).
- **\(b\)**: y-intercept (the point where the line crosses the y-axis).

  **Key Concept:**
  - The slope \(m\) is the ratio of the vertical change to the horizontal change between two points on the line.
  - The y-intercept \(b\) gives the value of \(y\) when \(x = 0\).

  **Example:**  
  If the equation is \(y = 2x + 3\), the slope is 2, and the line crosses the y-axis at \(y = 3\).

---

### 2. **Point-Slope Form**  
**Equation:**  
$$
y - y_1 = m(x - x_1)
$$  
- **\(m\)**: Slope of the line.
- **\((x_1, y_1)\)**: A specific point on the line.

  **Key Concept:**
  - This form is useful when you know a point on the line and the slope.
  - It shows how the line changes relative to any point \((x_1, y_1)\) and its slope \(m\).

  **Example:**  
  For a line with slope 3 passing through the point \((1, 4)\), the equation is:  
  $$
  y - 4 = 3(x - 1)
  $$

---

### 3. **Standard Form**  
**Equation:**  
$$
Ax + By = C
$$  
- **\(A\), \(B\), \(C\)**: Constants.
- \(A\) and \(B\) cannot both be zero.

  **Key Concept:**
  - This is a general form of a linear equation where \(x\) and \(y\) are the variables.
  - It can represent any line, but may require rearranging to other forms like slope-intercept for easier interpretation.

  **Example:**  
  If the equation is \(2x + 3y = 6\), this represents a line in standard form.

---

### 4. **General Form of a Line**  
**Equation:**  
$$
Ax + By + C = 0
$$  
- Similar to the standard form, but the constant \(C\) is on the right side of the equation.

  **Key Concept:**
  - This form is commonly used in analytical geometry and provides a straightforward way to work with lines in coordinate geometry.

  **Example:**  
  The equation \(4x - 5y + 10 = 0\) is a line in general form.

---

### 5. **Two-Point Form**  
**Equation:**  
$$
\frac{y - y_1}{y_2 - y_1} = \frac{x - x_1}{x_2 - x_1}
$$  
- **\((x_1, y_1)\)** and **\((x_2, y_2)\)** are two distinct points on the line.

  **Key Concept:**
  - This form is derived from the concept of the slope between two points \((x_1, y_1)\) and \((x_2, y_2)\).
  - It is useful when you know two points and need to find the equation of the line that passes through them.

  **Example:**  
  For the points \((1, 2)\) and \((3, 4)\), the equation is:
  $$
  \frac{y - 2}{4 - 2} = \frac{x - 1}{3 - 1}
  $$  
  Simplifying this gives the equation of the line.

---

### 6. **Horizontal and Vertical Lines**  
- **Horizontal Line:** The equation is of the form \(y = k\), where \(k\) is a constant.
  - This means that the value of \(y\) is the same for all points on the line.
  
  **Example:**  
  $$ y = 3 $$  
  This is a horizontal line passing through all points where \(y = 3\).

- **Vertical Line:** The equation is of the form \(x = h\), where \(h\) is a constant.
  - This means that the value of \(x\) is the same for all points on the line.

  **Example:**  
  $$ x = 2 $$  
  This is a vertical line passing through all points where \(x = 2\).

---

### 7. **Finding Slope from Two Points**  
To find the slope \(m\) of a line that passes through two points \((x_1, y_1)\) and \((x_2, y_2)\), use the formula:
$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$  
- The slope formula gives the "steepness" of the line.

---

### 8. **Perpendicular and Parallel Lines**  
- **Parallel Lines:** Two lines are parallel if they have the same slope.
  - If lines have slopes \(m_1\) and \(m_2\), then \(m_1 = m_2\) for parallel lines.
  
- **Perpendicular Lines:** Two lines are perpendicular if the product of their slopes is \(-1\).
  - If one line has slope \(m_1\), the other line must have slope \(m_2\) such that:
  $$
  m_1 \times m_2 = -1
  $$  
  - The slope of the line perpendicular to a line with slope \(m\) is \(-\frac{1}{m}\).

---

### 9. **Converting Between Forms**  
You can convert between different forms of the line equation by manipulating algebraically:

- From **Slope-Intercept to Standard Form:**  
  Start with \(y = mx + b\), and move all terms to one side:
  $$
  mx - y + b = 0
  $$

- From **Standard Form to Slope-Intercept:**  
  Solve for \(y\):
  $$
  y = \frac{-A}{B}x + \frac{C}{B}
  $$

---

### 10. **Distance from a Point to a Line**  
The formula for the distance \(d\) from a point \((x_0, y_0)\) to a line \(Ax + By + C = 0\) is:
$$
d = \frac{|Ax_0 + By_0 + C|}{\sqrt{A^2 + B^2}}
$$

---

- **y-intercept** is where the line crosses the y-axis.
- **Point-slope form** is useful when you know a point and the slope.
- **Standard and general forms** are more algebraic representations.
- **Parallel and perpendicular lines** have special relationships in terms of slopes.

Each of these forms and concepts provides a different way of analyzing and working with lines on the plane, and they are interconnected through basic algebraic manipulation.

The **parametric form of a line** is a way to represent a line using parameters, typically in two or three dimensions. It expresses the coordinates of any point on the line as a function of a parameter, often denoted as \( t \).

### **In 2D (two-dimensional space)**

A line in 2D can be represented by parametric equations in the form:

$$
\begin{cases}
x = x_0 + t \cdot v_x \\
y = y_0 + t \cdot v_y
\end{cases}
$$

Where:
- \( (x_0, y_0) \) is a point on the line (this could be any point through which the line passes),
- \( (v_x, v_y) \) is the direction vector of the line. This vector tells you the direction in which the line moves (how much \( x \) changes per unit of \( t \), and how much \( y \) changes per unit of \( t \)),
- \( t \) is the parameter, which represents the "time" or "step" at which you want to evaluate the position on the line.

### **How does the parametric form work?**
- For any value of \( t \), the parametric equations give you the coordinates \( (x(t), y(t)) \) of a point on the line.
- The parameter \( t \) can take any real value, and as it changes, it gives you all the points on the line.

### **Key Elements in the Parametric Form**:
1. **Point on the line**: \( (x_0, y_0) \) gives a starting point.
2. **Direction vector**: \( (v_x, v_y) \) gives the direction of movement.
   - For example, if \( v_x = 2 \) and \( v_y = 3 \), it means for every 1 unit you move horizontally (along \( x \)), you move 1.5 units vertically (along \( y \)).
3. **Parameter \( t \)**: The parameter allows you to trace all points on the line by varying its value.

### **Example**:

If you are given:
- A point \( A(1, 2) \),
- A direction vector \( \vec{v} = [2, 3] \) (i.e., for every 2 units moved in the \( x \)-direction, 3 units are moved in the \( y \)-direction),

The parametric equations for the line passing through \( A(1, 2) \) and parallel to \( \vec{v} \) would be:

$$
\begin{cases}
x(t) = 1 + 2t \\
y(t) = 2 + 3t
\end{cases}
$$

For different values of \( t \), you get different points on the line:
- When \( t = 0 \), the point is \( (1, 2) \),
- When \( t = 1 \), the point is \( (3, 5) \),
- When \( t = -1 \), the point is \( (-1, -1) \), and so on.

### **In 3D (three-dimensional space)**

For a line in 3D, the parametric form becomes:

$$
\begin{cases}
x = x_0 + t \cdot v_x \\
y = y_0 + t \cdot v_y \\
z = z_0 + t \cdot v_z
\end{cases}
$$

Where:
- \( (x_0, y_0, z_0) \) is a point on the line,
- \( (v_x, v_y, v_z) \) is the direction vector of the line in three dimensions,
- \( t \) is the parameter.

In this case, by changing \( t \), you get different points on the line in 3D space.

### **Summary**:
The parametric form of a line is a way of describing all the points on the line by using a parameter \( t \), which changes to generate different points. The line's direction and a point on the line are key components in determining the parametric equations.

$$

- \( (x_0, y_0) \) is the point through which the line passes (here \( A(1, 2) \)),

$$
\begin{cases}

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parametric equations
def line_parametric(t):
    x = 1 + 2 * t
    y = 2 + 3 * t
    return x, y

# Generate t values to plot the line
t_values = np.linspace(-10, 10, 400)
x_values, y_values = line_parametric(t_values)

# Plot the line and the point A(1, 2)
plt.figure(figsize=(6,6))
plt.plot(x_values, y_values, label=r'Line passing through $A(1, 2)$ and parallel to $\vec{v} = [2, 3]$', color='blue')
plt.scatter(1, 2, color='red', label='Point A(1, 2)', zorder=5)

# Adding labels and title
plt.title('Line passing through A(1,2) and parallel to vector v=[2,3]')
plt.xlabel('x')
plt.ylabel('y')

# Set the same scale for x and y axes
plt.axis('equal')

# Display grid
plt.grid(True)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the line equation y = -3/2 * x + 3
def line_equation(x):
    return -3/2 * x + 3

# Generate x values for the line
x = np.linspace(-1, 3, 100)
y = line_equation(x)

# Points A(2, 0) and B(0, 3)
A = (2, 0)
B = (0, 3)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r"$y = -\frac{3}{2}x + 3$", color='blue', linewidth=2)
plt.scatter(*A, color='red', label='Point A(2, 0)', zorder=5)
plt.scatter(*B, color='green', label='Point B(0, 3)', zorder=5)

# Add annotations
plt.text(A[0], A[1] - 0.2, "A(2, 0)", color='red', fontsize=10, ha='center')
plt.text(B[0] - 0.3, B[1], "B(0, 3)", color='green', fontsize=10, ha='center')

# Axes and grid
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.title("Line Passing Through Points A(2, 0) and B(0, 3)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim(-1, 3)
plt.ylim(-1, 4)
plt.show()
```
