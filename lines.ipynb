### Equations of Lines on a Plane

In geometry, the equation of a line on a 2D Cartesian plane describes all the points that lie on that line. The equation of a line can be represented in various forms, depending on the given information about the line.

---

### 1. **Slope-Intercept Form**  
**Equation:**  
$$
y = mx + b
$$  
- **\(m\)**: Slope of the line (rise over run, or the rate of change of \(y\) with respect to \(x\)).
- **\(b\)**: y-intercept (the point where the line crosses the y-axis).

  **Key Concept:**
  - The slope \(m\) is the ratio of the vertical change to the horizontal change between two points on the line.
  - The y-intercept \(b\) gives the value of \(y\) when \(x = 0\).

  **Example:**  
  If the equation is \(y = 2x + 3\), the slope is 2, and the line crosses the y-axis at \(y = 3\).

---

### 2. **Point-Slope Form**  
**Equation:**  
$$
y - y_1 = m(x - x_1)
$$  
- **\(m\)**: Slope of the line.
- **\((x_1, y_1)\)**: A specific point on the line.

  **Key Concept:**
  - This form is useful when you know a point on the line and the slope.
  - It shows how the line changes relative to any point \((x_1, y_1)\) and its slope \(m\).

  **Example:**  
  For a line with slope 3 passing through the point \((1, 4)\), the equation is:  
  $$
  y - 4 = 3(x - 1)
  $$

---

### 3. **Standard Form**  
**Equation:**  
$$
Ax + By = C
$$  
- **\(A\), \(B\), \(C\)**: Constants.
- \(A\) and \(B\) cannot both be zero.

  **Key Concept:**
  - This is a general form of a linear equation where \(x\) and \(y\) are the variables.
  - It can represent any line, but may require rearranging to other forms like slope-intercept for easier interpretation.

  **Example:**  
  If the equation is \(2x + 3y = 6\), this represents a line in standard form.

---

### 4. **General Form of a Line**  
**Equation:**  
$$
Ax + By + C = 0
$$  
- Similar to the standard form, but the constant \(C\) is on the right side of the equation.

  **Key Concept:**
  - This form is commonly used in analytical geometry and provides a straightforward way to work with lines in coordinate geometry.

  **Example:**  
  The equation \(4x - 5y + 10 = 0\) is a line in general form.

---

### 5. **Two-Point Form**  
**Equation:**  
$$
\frac{y - y_1}{y_2 - y_1} = \frac{x - x_1}{x_2 - x_1}
$$  
- **\((x_1, y_1)\)** and **\((x_2, y_2)\)** are two distinct points on the line.

  **Key Concept:**
  - This form is derived from the concept of the slope between two points \((x_1, y_1)\) and \((x_2, y_2)\).
  - It is useful when you know two points and need to find the equation of the line that passes through them.

  **Example:**  
  For the points \((1, 2)\) and \((3, 4)\), the equation is:
  $$
  \frac{y - 2}{4 - 2} = \frac{x - 1}{3 - 1}
  $$  
  Simplifying this gives the equation of the line.

---

### 6. **Horizontal and Vertical Lines**  
- **Horizontal Line:** The equation is of the form \(y = k\), where \(k\) is a constant.
  - This means that the value of \(y\) is the same for all points on the line.
  
  **Example:**  
  $$ y = 3 $$  
  This is a horizontal line passing through all points where \(y = 3\).

- **Vertical Line:** The equation is of the form \(x = h\), where \(h\) is a constant.
  - This means that the value of \(x\) is the same for all points on the line.

  **Example:**  
  $$ x = 2 $$  
  This is a vertical line passing through all points where \(x = 2\).

---

### 7. **Finding Slope from Two Points**  
To find the slope \(m\) of a line that passes through two points \((x_1, y_1)\) and \((x_2, y_2)\), use the formula:
$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$  
- The slope formula gives the "steepness" of the line.

---

### 8. **Perpendicular and Parallel Lines**  
- **Parallel Lines:** Two lines are parallel if they have the same slope.
  - If lines have slopes \(m_1\) and \(m_2\), then \(m_1 = m_2\) for parallel lines.
  
- **Perpendicular Lines:** Two lines are perpendicular if the product of their slopes is \(-1\).
  - If one line has slope \(m_1\), the other line must have slope \(m_2\) such that:
  $$
  m_1 \times m_2 = -1
  $$  
  - The slope of the line perpendicular to a line with slope \(m\) is \(-\frac{1}{m}\).

---

### 9. **Converting Between Forms**  
You can convert between different forms of the line equation by manipulating algebraically:

- From **Slope-Intercept to Standard Form:**  
  Start with \(y = mx + b\), and move all terms to one side:
  $$
  mx - y + b = 0
  $$

- From **Standard Form to Slope-Intercept:**  
  Solve for \(y\):
  $$
  y = \frac{-A}{B}x + \frac{C}{B}
  $$

---

### 10. **Distance from a Point to a Line**  
The formula for the distance \(d\) from a point \((x_0, y_0)\) to a line \(Ax + By + C = 0\) is:
$$
d = \frac{|Ax_0 + By_0 + C|}{\sqrt{A^2 + B^2}}
$$

---

### Summary of Key Concepts:
- **Slope** indicates how steep the line is.
- **y-intercept** is where the line crosses the y-axis.
- **Point-slope form** is useful when you know a point and the slope.
- **Standard and general forms** are more algebraic representations.
- **Parallel and perpendicular lines** have special relationships in terms of slopes.

Each of these forms and concepts provides a different way of analyzing and working with lines on the plane, and they are interconnected through basic algebraic manipulation.
