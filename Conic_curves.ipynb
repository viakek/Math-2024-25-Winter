### Notes on Conic Sections

#### **1. Equation of a Circle**
- **Standard Form**: The equation of a circle with center \( (h, k) \) and radius \( r \) is:
  $$
  (x-h)^2 + (y-k)^2 = r^2
  $$
- **Key Points**:
  - \( h, k \) define the center.
  - \( r \) is the radius.
- **Example**: For a circle with center \( A(1, 2) \) and \( r = 3 \), the equation becomes:
  $$
  (x-1)^2 + (y-2)^2 = 9
  $$

---

#### **2. Equation of a Parabola**
- **General Form**:
  - If a parabola intersects the \( O_x \)-axis at points \( x = p \) and \( x = q \), its equation can be expressed as:
    $$
    y = a(x-p)(x-q)
    $$
  - The coefficient \( a \) is determined using a third point on the parabola.
- **Example**:
  - Given intersections at \( x = 2 \) and \( x = 4 \), the equation is:
    $$
    y = a(x-2)(x-4)
    $$
  - Use the point \( (3, 1) \) to solve for \( a \):
    $$
    1 = a(3-2)(3-4) \implies a = -1
    $$
  - Final Equation:
    $$
    y = -(x-2)(x-4)
    $$

---

#### **3. Center of an Ellipse**
- **Standard Form of an Ellipse**:
  $$
  Ax^2 + By^2 + Dx + Ey + F = 0
  $$
- **Completing the Square**:
  - Group terms for \( x \) and \( y \), and complete the square.
  - Example: For \( x^2 + 4y^2 - 4x - 16y + 16 = 0 \):
    - Rearrange: \( (x^2 - 4x) + 4(y^2 - 4y) = -16 \)
    - Complete squares:
      $$
      (x-2)^2 - 4 + 4((y-2)^2 - 4) = -16
      $$
      $$
      (x-2)^2 + 4(y-2)^2 = 4
      $$
    - Final Form:
      $$
      \frac{(x-2)^2}{4} + \frac{(y-2)^2}{1} = 1
      $$
    - Center: \( (2, 2) \).

---

#### **4. Slope of a Tangent Line to a Circle**
- **Circle Equation**:
  $$
  x^2 + y^2 = r^2
  $$
- **Condition for Tangency**:
  - A line \( y = mx + c \) is tangent to a circle if the discriminant of the quadratic equation formed by substituting \( y = mx + c \) into the circle equation equals 0.
  - For \( y = mx - 5 \) and \( x^2 + y^2 = 1 \):
    - Substitute \( y = mx - 5 \):
      $$
      x^2 + (mx - 5)^2 = 1
      $$
      $$
      x^2 + m^2x^2 - 10mx + 25 = 1
      $$
      $$
      (1+m^2)x^2 - 10mx + 24 = 0
      $$
    - Discriminant:
      $$
      \Delta = (-10m)^2 - 4(1+m^2)(24) = 0
      $$
      Solve for \( m \).

---

#### **5. Intersection of a Hyperbola and an Ellipse**
- **Hyperbola**:
  $$
  x^2 - y^2 = 1
  $$
- **Ellipse**:
  $$
  x^2 + 4y^2 = 6
  $$
- **Solve Simultaneously**:
  - Substitute \( x^2 = 1 + y^2 \) (from the hyperbola) into the ellipse equation:
    $$
    1 + y^2 + 4y^2 = 6
    $$
    $$
    5y^2 = 5 \implies y^2 = 1 \implies y = \pm 1
    $$
  - Solve for \( x \) using \( x^2 = 1 + y^2 \):
    $$
    x^2 = 2 \implies x = \pm \sqrt{2}
    $$
  - Intersection Points:
    $$
    (\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, 1), (-\sqrt{2}, -1)
    $$

#### **6. Distance Between Branches of a Hyperbola**

### Notes on the Distance Between the Branches of a Hyperbola

A hyperbola is defined as the set of all points \((x, y)\) such that the difference in distances to two fixed points (called foci) is constant. The general equation of a hyperbola centered at the origin can take two standard forms:

1. **Horizontal transverse axis**:
   $$
   \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
   $$

2. **Vertical transverse axis**:
   $$
   \frac{y^2}{a^2} - \frac{x^2}{b^2} = 1
   $$

### Parameters:
- **\( a \)**: The distance from the center to each vertex (along the transverse axis).
- **\( b \)**: The distance from the center to each co-vertex (along the conjugate axis).
- **\( c \)**: The distance from the center to each focus, where \( c^2 = a^2 + b^2 \).

### Distance Between the Branches:
The distance between the branches of a hyperbola refers to the distance between the two branches along the line perpendicular to the transverse axis (typically along the y-axis for a horizontally oriented hyperbola). This distance can be calculated using the formula:

$$
\text{Distance} = \frac{2b^2}{a}
$$

### Derivation of the Formula:
1. **For a hyperbola centered at the origin** with a horizontal transverse axis, the equation is:
   $$
   \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
   $$

2. **Asymptotes of the hyperbola**:
   The asymptotes are lines that pass through the center and define the directions in which the branches extend. For a horizontally oriented hyperbola, the equations of the asymptotes are:
   $$
   y = \pm \frac{b}{a} x
   $$

3. **Distance between branches**:
   The distance between the branches is the vertical distance between the points on the y-axis where the asymptotes intersect. This distance is given by:
   $$
   \text{Distance} = \frac{2b^2}{a}
   $$

### Example: Hyperbola \( x^2 - y^2 = 1 \)
Given the equation \( x^2 - y^2 = 1 \):

- \( a^2 = 1 \), so \( a = 1 \)
- \( b^2 = 1 \), so \( b = 1 \)

Using the distance formula:
$$
\text{Distance} = \frac{2b^2}{a} = \frac{2 \cdot 1}{1} = 2
$$

### Final Answer:
The distance between the branches of the hyperbola \( x^2 - y^2 = 1 \) is:
$$
\text{Distance} = 2
$$

### Important Notes:
- This distance is only relevant for hyperbolas with a horizontal transverse axis. For vertical hyperbolas, a similar formula \( \frac{2a^2}{b} \) would be used.
- The branches of the hyperbola extend infinitely, but the distance calculated refers to the distance at the intersection of the asymptotes with the conjugate axis.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the circle
def plot_circle():
    theta = np.linspace(0, 2 * np.pi, 100)  # Parameter for angles
    x = 1 + 3 * np.cos(theta)  # Circle's x-coordinates
    y = 2 + 3 * np.sin(theta)  # Circle's y-coordinates

    plt.figure(figsize=(6, 6))
    plt.plot(x, y, label=r"Circle: $(x-1)^2 + (y-2)^2 = 9$", color="blue")
    plt.scatter(1, 2, color="red", label="Center: (1,2)")
    plt.gca().set_aspect('equal', adjustable='box')  # Equal aspect ratio
    plt.axhline(0, color="black", linewidth=0.5)
    plt.axvline(0, color="black", linewidth=0.5)
    plt.title("Circle Visualization")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.legend()
    plt.grid()
    plt.show()

# Call the function to plot the circle
plot_circle()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parabola equation
def plot_parabola():
    # Find the parabola equation using vertex form: y = a(x - x1)(x - x2)
    x1, x2 = 2, 4  # x-intercepts
    point = (3, 1)  # Point (3, 1) lies on the parabola

    # Solve for 'a' using y = a(x - x1)(x - x2)
    x, y = point
    a = y / ((x - x1) * (x - x2))  # Calculate 'a'

    # Generate x values for plotting
    x_vals = np.linspace(1, 5, 500)
    y_vals = a * (x_vals - x1) * (x_vals - x2)  # Calculate y values

    # Plot the parabola
    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label=f"Parabola: $y = {a:.2f}(x - {x1})(x - {x2})$", color="blue")
    plt.scatter([x1, x2], [0, 0], color="red", label=f"Intercepts: ({x1}, 0), ({x2}, 0)")
    plt.scatter(x, y, color="green", label=f"Point: ({x}, {y})")
    plt.axhline(0, color="black", linewidth=0.5)
    plt.axvline(0, color="black", linewidth=0.5)
    plt.title("Parabola Visualization")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.legend()
    plt.grid()
    plt.show()

# Call the function to plot the parabola
plot_parabola()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_ellipse():
    # Rewrite the given equation in standard form
    # Given: x^2 + 4y^2 - 4x - 16y + 16 = 0
    # Completing the square for x and y terms:
    # x^2 - 4x -> (x - 2)^2 - 4
    # 4y^2 - 16y -> 4((y - 2)^2 - 4)
    # Substituting back: (x - 2)^2 + 4(y - 2)^2 = 4 + 16 - 16 = 4
    # Final equation: (x - 2)^2 / 4 + (y - 2)^2 / 1 = 1

    # Ellipse parameters
    h, k = 2, 2  # Center of the ellipse (2, 2)
    a, b = 2, 1  # Semi-major axis (a) and semi-minor axis (b)

    # Generate points for the ellipse
    theta = np.linspace(0, 2 * np.pi, 500)
    x = h + a * np.cos(theta)
    y = k + b * np.sin(theta)

    # Plot the ellipse
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label="Ellipse: $(x - 2)^2 / 4 + (y - 2)^2 / 1 = 1$", color="blue")
    plt.scatter(h, k, color="red", label=f"Center: ({h}, {k})")
    plt.axhline(0, color="black", linewidth=0.5)
    plt.axvline(0, color="black", linewidth=0.5)
    plt.title("Ellipse Visualization")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.legend()
    plt.grid()
    plt.axis("equal")
    plt.show()

# Call the function to plot the ellipse
plot_ellipse()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_circle_and_tangent():
    # Circle parameters
    r = 1  # Radius of the circle
    h, k = 0, 0  # Center of the circle at origin (0, 0)

    # Tangent line parameters
    m = 5 / np.sqrt(26)  # Slope of the tangent
    c = -5  # y-intercept of the tangent line
    tangent_x = np.linspace(-1.5, 1.5, 500)
    tangent_y = m * tangent_x + c

    # Circle equation: x^2 + y^2 = r^2
    theta = np.linspace(0, 2 * np.pi, 500)
    circle_x = r * np.cos(theta)
    circle_y = r * np.sin(theta)

    # Plot the circle
    plt.figure(figsize=(8, 8))
    plt.plot(circle_x, circle_y, label="Circle: $x^2 + y^2 = 1$", color="blue")
    plt.axhline(0, color="black", linewidth=0.5)
    plt.axvline(0, color="black", linewidth=0.5)

    # Plot the tangent line
    plt.plot(tangent_x, tangent_y, label=f"Tangent line: $y = {m:.2f}x - 5$", color="red", linestyle="--")

    # Plot the tangent point
    tangent_point_x = 1 / np.sqrt(26)
    tangent_point_y = 5 / np.sqrt(26)
    plt.scatter(tangent_point_x, tangent_point_y, color="green", label=f"Tangent point: ({tangent_point_x:.2f}, {tangent_point_y:.2f})")

    # Set up the graph
    plt.title("Circle and Tangent Line Visualization")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.legend()
    plt.grid()
    plt.axis("equal")
    plt.show()

# Call the function to plot the circle and tangent
plot_circle_and_tangent()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve

def plot_hyperbola_and_ellipse():
    # Define the grid
    x = np.linspace(-3, 3, 500)
    y = np.linspace(-3, 3, 500)
    X, Y = np.meshgrid(x, y)

    # Equations of the hyperbola and the ellipse
    hyperbola = X**2 - Y**2 - 1
    ellipse = X**2 + 4 * Y**2 - 6

    # Plot hyperbola
    plt.contour(X, Y, hyperbola, levels=[0], colors="blue", label="Hyperbola: $x^2 - y^2 = 1$")

    # Plot ellipse
    plt.contour(X, Y, ellipse, levels=[0], colors="red", label="Ellipse: $x^2 + 4y^2 = 6$")

    # Solve for intersection points symbolically
    x_sym, y_sym = symbols('x y')
    hyperbola_eq = Eq(x_sym**2 - y_sym**2, 1)
    ellipse_eq = Eq(x_sym**2 + 4 * y_sym**2, 6)
    intersection_points = solve((hyperbola_eq, ellipse_eq), (x_sym, y_sym))

    # Extract intersection points
    for point in intersection_points:
        plt.scatter(float(point[0]), float(point[1]), color="green", label=f"Intersection: {point}")

    # Labels and visualization settings
    plt.title("Intersection of Hyperbola and Ellipse")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.axhline(0, color="black", linewidth=0.5)
    plt.axvline(0, color="black", linewidth=0.5)
    plt.legend(["Hyperbola: $x^2 - y^2 = 1$", "Ellipse: $x^2 + 4y^2 = 6$"])
    plt.grid()
    plt.axis("equal")
    plt.show()

# Call the function to plot the hyperbola and ellipse
plot_hyperbola_and_ellipse()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_hyperbola_and_distance():
    # Define the grid
    x = np.linspace(-3, 3, 500)
    y = np.linspace(-3, 3, 500)
    X, Y = np.meshgrid(x, y)

    # Equation of the hyperbola
    hyperbola = X**2 - Y**2 - 1

    # Plot the hyperbola
    plt.contour(X, Y, hyperbola, levels=[0], colors="blue", label="Hyperbola: $x^2 - y^2 = 1$")

    # Branches of the hyperbola are symmetric about the y-axis
    # Minimum horizontal distance between branches is along the x-axis at y = 0
    x_branch1 = 1  # Right branch (x = sqrt(1 + y^2), y = 0)
    x_branch2 = -1  # Left branch (x = -sqrt(1 + y^2), y = 0)
    y_intersect = 0  # Along the x-axis

    # Highlight the distance between the branches
    plt.plot([x_branch1, x_branch2], [y_intersect, y_intersect], 'ro-')  # Distance line
    plt.text(0, -0.5, "Distance = 2", color="red", fontsize=12, ha="center")  # Annotate the distance

    # Labels and visualization settings
    plt.title("Distance Between Branches of Hyperbola")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.axhline(0, color="black", linewidth=0.5)
    plt.axvline(0, color="black", linewidth=0.5)
    plt.grid()
    plt.axis("equal")
    plt.legend(["Hyperbola: $x^2 - y^2 = 1$", "Branch Distance"])
    plt.show()

# Call the function to plot the hyperbola and its branch distance
plot_hyperbola_and_distance()
```
