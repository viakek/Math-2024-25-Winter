# Vectors - Projections and Divisions


## Projecting a Vector onto Another Vector

Projecting vector $\mathbf{a}$ onto axis $\mathbf{b}$ involves finding the component of vector $\mathbf{a}$ along the direction of $\mathbf{b}$.

Projection of $\mathbf{a}$ onto $\mathbf{b}$:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{|\mathbf{b}|^2} \mathbf{b}
$$


## Projecting a Vector onto Axes

Projecting vector $\mathbf{a}$ onto an axis involves finding the component of vector $\mathbf{a}$ along that axis.

Projection of $\mathbf{a}$ onto the $OX$ axis:

$$
\text{Proj}_{OX} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{i}}{|\mathbf{i}|^2} \mathbf{i} = \frac{\mathbf{a} \cdot \mathbf{i}}{1} \mathbf{i} = (\mathbf{a} \cdot \mathbf{i}) \mathbf{i}
$$

Projection of $\mathbf{a}$ onto the $OY$ axis:

$$
\text{Proj}_{OY} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{j}}{|\mathbf{j}|^2} \mathbf{j} = \frac{\mathbf{a} \cdot \mathbf{j}}{1} \mathbf{j} = (\mathbf{a} \cdot \mathbf{j}) \mathbf{j}
$$

Recall that $\mathbf{i} = (1, 0)$ and $\mathbf{j} = (0, 1)$ are unit vectors along the $OX$ and $OY$ axes.


## Dividing a Line Segment in a Given Ratio

A line segment is defined by two points in space.

Line segments and vectors are closely related because a line segment connecting two points can be treated as a vector directed from one point to another.

If point \( P \) divides the line segment connecting points $A(x_1, y_1)$ and $B(x_2, y_2)$ in the ratio $m:n$, then the coordinates of point $P$ can be found using the formula:

$$
P = \left(\frac{n \times x_2 + m \times x_1}{m+n}, \frac{n \times y_2 + m \times y_1}{m+n}\right)
$$

**Example:**

Given a line segment connecting points $A(1, 2)$ and $B(3, 8)$, find point $P$ that divides the segment in the ratio 2:3.

Coordinates of point $P$:
$$
P = \left(\frac{2 \times 3 + 3 \times 1}{2+3}, \frac{2 \times 8 + 3 \times 2}{2+3}\right) = \left(\frac{9}{5}, \frac{22}{5}\right)
$$

```python
import matplotlib.pyplot as plt
import math

# Coordinates of points A and B
A = (1, 2)
B = (3, 8)

# Ratio m:n
m, n = 2, 3

# Calculate coordinates of point P that divides the segment in the ratio 2:3
P_x = (n * A[0] + m * B[0]) / (m + n)
P_y = (n * A[1] + m * B[1]) / (m + n)
P = (P_x, P_y)

# Plotting the graph
plt.figure(figsize=(6, 6))

# Plotting line segment AB
plt.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Line Segment AB')

# Marking points A, B, and P
plt.scatter(*A, color='blue', label='A(1, 2)')
plt.scatter(*B, color='red', label='B(3, 8)')
plt.scatter(*P, color='green', label=f'P({P[0]:.2f}, {P[1]:.2f})')

# Point labels
plt.text(A[0]-0.2, A[1], 'A(1, 2)', fontsize=12, color='blue')
plt.text(B[0]+0.1, B[1], 'B(3, 8)', fontsize=12, color='red')
plt.text(P[0]+0.1, P[1], f'P({P[0]:.2f}, {P[1]:.2f})', fontsize=12, color='green')

# Adding grid and axis labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dividing a Line Segment in the Ratio 2:3')
plt.legend()

# Displaying the graph
plt.show()
```

## Student Exercises

- Divide the line segment connecting points $A(-1, 2)$ and $B(3, -2)$ in the ratio $1:3$. Illustrate the result on a graph.

### **Solution**

$$
P(x, y) = \left( \frac{m x_2 + n x_1}{m + n}, \frac{m y_2 + n y_1}{m + n} \right) \\
$$


$$  A(x_1, y_1) = (-1, 2) \\
 B(x_2, y_2) = (3, -2) \\
 m = 1 \\  n = 3  $$

#### **Finding \( x \):**
$$
x = \frac{m x_2 + n x_1}{m + n} = \frac{1 \cdot 3 + 3 \cdot (-1)}{1 + 3} = \frac{3 - 3}{4} = 0
$$

#### **Finding \( y \):**
$$
y = \frac{m y_2 + n y_1}{m + n} = \frac{1 \cdot (-2) + 3 \cdot 2}{1 + 3} = \frac{-2 + 6}{4} = 1
$$

the point \( P \) is:
$$
P(0, 1)
$$

- Project vector $\mathbf{a} = (2,3)$ onto vector $\mathbf{b} = (1, 1)$. Illustrate the result on a graph.

### **Solution**


$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{\mathbf{b} \cdot \mathbf{b}} \mathbf{b}
$$


$$
\mathbf{a} \cdot \mathbf{b} = (2)(1) + (3)(1) = 2 + 3 = 5
$$

$$
\mathbf{b} \cdot \mathbf{b} = (1)^2 + (1)^2 = 1 + 1 = 2
$$

$$
\frac{\mathbf{a} \cdot \mathbf{b}}{\mathbf{b} \cdot \mathbf{b}} = \frac{5}{2}
$$

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{5}{2} \mathbf{b} \\ = \frac{5}{2} (1, 1) = \left( \frac{5}{2}, \frac{5}{2} \right) \\ = (2.5, 2.5)
$$


- Project vector $\mathbf{b} = (1, 1)$ onto vector $\mathbf{a} = (2, 3)$. Illustrate the result on a graph.

### **Solution**

$$
\text{Proj}_{\mathbf{a}} \mathbf{b} = \frac{\mathbf{b} \cdot \mathbf{a}}{\mathbf{a} \cdot \mathbf{a}} \mathbf{a}
$$

$$
\mathbf{b} \cdot \mathbf{a} = (1)(2) + (1)(3) = 2 + 3 = 5
$$


$$
\mathbf{a} \cdot \mathbf{a} = (2)^2 + (3)^2 = 4 + 9 = 13
$$


$$
\frac{\mathbf{b} \cdot \mathbf{a}}{\mathbf{a} \cdot \mathbf{a}} = \frac{5}{13}
$$

$$
\text{Proj}_{\mathbf{a}} \mathbf{b} = \frac{5}{13} \mathbf{a} = \frac{5}{13} (2, 3) \\ = \left( \frac{10}{13}, \frac{15}{13} \right) \\
$$



```python
import matplotlib.pyplot as plt

# Coordinates of points A and B
A = (-1, 2)
B = (3, -2)

# Ratio 1:3
m, n = 1, 3

# Using the section formula to find the dividing point
x = (m * B[0] + n * A[0]) / (m + n)
y = (m * B[1] + n * A[1]) / (m + n)
P = (x, y)  # Point dividing the segment in the ratio 1:3

# Plotting the points and line segment
plt.figure(figsize=(6, 6))
plt.plot([A[0], B[0]], [A[1], B[1]], 'b-', label="Line segment AB")
plt.scatter(*A, color='red', label='A (-1, 2)', zorder=5)
plt.scatter(*B, color='green', label='B (3, -2)', zorder=5)
plt.scatter(*P, color='orange', label=f'P {P}', zorder=5)

# Adding labels and grid
plt.text(A[0] - 0.5, A[1] + 0.2, 'A', fontsize=10)
plt.text(B[0] + 0.2, B[1] - 0.5, 'B', fontsize=10)
plt.text(P[0] + 0.2, P[1] + 0.2, 'P', fontsize=10)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.title("Division of Line Segment AB in Ratio 1:3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Vector a
a = np.array([3, 4])

# Projections
a_ox = np.array([3, 0])  # Projection onto OX (x-axis)
a_oy = np.array([0, 4])  # Projection onto OY (y-axis)

# Origin
origin = np.array([0, 0])

# Plotting
plt.figure(figsize=(8, 8))
plt.quiver(*origin, *a, color='blue', angles='xy', scale_units='xy', scale=1, label='Vector a=(3,4)')
plt.quiver(*origin, *a_ox, color='red', angles='xy', scale_units='xy', scale=1, label='Projection on OX')
plt.quiver(*origin, *a_oy, color='green', angles='xy', scale_units='xy', scale=1, label='Projection on OY')

# Grid, labels, and legend
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.legend()
plt.title("Projection of Vector a=(3,4) onto OX and OY Axes")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Vectors a and b
a = np.array([2, 3])
b = np.array([1, 1])

# Calculate projection of a onto b
b_magnitude_squared = np.dot(b, b)
projection_scalar = np.dot(a, b) / b_magnitude_squared
projection = projection_scalar * b

# Origin
origin = np.array([0, 0])

# Plotting
plt.figure(figsize=(8, 8))
plt.quiver(*origin, *a, color='blue', angles='xy', scale_units='xy', scale=1, label='Vector a=(2,3)')
plt.quiver(*origin, *b, color='green', angles='xy', scale_units='xy', scale=1, label='Vector b=(1,1)')
plt.quiver(*origin, *projection, color='red', angles='xy', scale_units='xy', scale=1, label='Projection of a onto b')

# Grid, labels, and legend
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.legend()
plt.title("Projection of Vector a=(2,3) onto Vector b=(1,1)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Vectors a and b
a = np.array([2, 3])
b = np.array([1, 1])

# Calculate projection of b onto a
a_magnitude_squared = np.dot(a, a)
projection_scalar = np.dot(b, a) / a_magnitude_squared
projection = projection_scalar * a

# Origin
origin = np.array([0, 0])

# Plotting
plt.figure(figsize=(8, 8))
plt.quiver(*origin, *a, color='blue', angles='xy', scale_units='xy', scale=1, label='Vector a=(2,3)')
plt.quiver(*origin, *b, color='green', angles='xy', scale_units='xy', scale=1, label='Vector b=(1,1)')
plt.quiver(*origin, *projection, color='red', angles='xy', scale_units='xy', scale=1, label='Projection of b onto a')

# Grid, labels, and legend
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.legend()
plt.title("Projection of Vector b=(1,1) onto Vector a=(2,3)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
