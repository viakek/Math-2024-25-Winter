$$\textbf{Notes on Equations of Planes in Space}$$

## 1. Plane Through Three Points

**Task:** Find the plane passing through $A(1,2,3)$, $B(3,4,5)$, and $C(2,1,4)$.  

**Notes:**  
- Compute two vectors on the plane:  
  $$\vec{AB} = \vec{B} - \vec{A}, \quad \vec{AC} = \vec{C} - \vec{A}.$$
- Find the normal vector using the cross product:  
  $$\vec{n} = \vec{AB} \times \vec{AC}.$$
- The equation of the plane is:  
  $$a(x - x_1) + b(y - y_1) + c(z - z_1) = 0,$$  
  where $(a, b, c)$ is the normal vector and $(x_1, y_1, z_1)$ is a point on the plane.

---

## 2. Plane Through a Point and Parallel to Another Plane

**Task:** Find the plane passing through $A(1,2,3)$ and parallel to $2x+3y+4z=5$.  

**Notes:**  
- Parallel planes have the same normal vector.  
  $$\vec{n} = [2, 3, 4].$$
- The general form is:  
  $$2x + 3y + 4z + D = 0.$$
- Use the point $A(1,2,3)$ to solve for $D$.

---

## 3. Plane Through a Point and Perpendicular to a Vector

**Task:** Find the plane passing through $A(1,2,3)$ and perpendicular to $\vec{n} = [2, 3, 4]$.  

**Notes:**  
- Use the point-normal form of the plane:  
  $$2(x - 1) + 3(y - 2) + 4(z - 3) = 0.$$

---

## 4. Line of Intersection of Two Planes

**Task:** Find the line of intersection of $2x + 3y + 4z = 5$ and $3x + 4y + 2z = 6$.  

**Notes:**  
- Solve the system of equations to eliminate one variable (e.g., $z$).  
- Parametric form of the line:  
  $$x = x_0 + t, \quad y = y_0 + t, \quad z = z_0 + t.$$

---

## 5. Plane Through a Point and Parallel to Two Vectors

**Task:** Find the plane passing through $A(1,2,3)$ and parallel to $\vec{v}_1 = [1,0,1]$ and $\vec{v}_2 = [0,1,-1]$.  

**Notes:**  
- Compute the normal vector:  
  $$\vec{n} = \vec{v}_1 \times \vec{v}_2.$$
- Use the point-normal form:  
  $$a(x - x_1) + b(y - y_1) + c(z - z_1) = 0.$$

---

## 6. Parallel and Perpendicular Planes

**Task:** Find planes parallel and perpendicular to $2x + 3y + 4z = 5$.  

**Notes:**  
- Parallel planes: Keep the same normal vector, adjust $D$:  
  $$2x + 3y + 4z + D = 0.$$  
- Perpendicular planes: Choose a normal vector perpendicular to $[2,3,4]$, such as $[4,-2,3]$, and form the equation.

---

## 7. Distance from a Point to a Plane

**Task:** Find the distance from $A(1,2,3)$ to the plane $2x + 3y + 4z = 5$.  

**Formula:**  
$$d = \frac{|ax_1 + by_1 + cz_1 + D|}{\sqrt{a^2 + b^2 + c^2}}.$$

**Notes:**  
- Substitute $A(1,2,3)$ into the formula with $a=2, b=3, c=4, D=-5$.

---

## 8. Plane Through Three Points of Intersection

**Task:** Find the plane passing through $A(2,0,0)$, $B(0,3,0)$, and $C(0,0,4)$.  

**Notes:**  
- Follow the steps in **Task 1** to compute the normal vector and form the plane equation.

---

## 9. Angle Between Two Planes

**Task:** Find the angle between $x + y + z = 1$ and the $yz$-plane $(x=0)$.  

**Formula:**  
$$\cos \theta = \frac{|\vec{n}_1 \cdot \vec{n}_2|}{\|\vec{n}_1\| \|\vec{n}_2\|}.$$

**Notes:**  
- Normal vector of $x + y + z = 1$ is $[1,1,1]$.  
- Normal vector of $x = 0$ (the $yz$-plane) is $[1,0,0]$.  

---

## 10. Perpendicular Vector to a Plane

**Task:** Find a vector perpendicular to $x + y + z = 1$.  

**Notes:**  
- The normal vector of the plane, $[1,1,1]$, is the required perpendicular vector.

---

## Summary

- Use the point-normal form for planes passing through a point and having a normal vector.  
- Use cross products to compute normal vectors for planes defined by points or parallel vectors.  
- Distances and angles involve formulas using vector norms and dot products.

* Find the vector perpendicular to the plane $x + y + z = 1$.
###Answer+notes on this question----->
### Identify the normal vector:

$$
\mathbf{n} = (1, 1, 1)
$$


### Explanation:
The normal vector \(\mathbf{n} = (1, 1, 1)\) is perpendicular to every vector that lies within the plane. This means that if you take any two vectors that lie on the plane and take their dot product with \(\mathbf{n}\), the result will be zero, confirming that the vectors are orthogonal.

### Conclusion:
The normal vector to the plane \(x + y + z = 1\) is:
$$
\mathbf{n} = (1, 1, 1)
$$
This vector points perpendicularly out from the plane, and it is orthogonal to any vector lying within the plane.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define points A, B, and C
A = np.array([1, 2, 3])
B = np.array([3, 4, 5])
C = np.array([2, 1, 4])

# Calculate vectors AB and AC
AB = B - A
AC = C - A

# Calculate the normal vector by taking the cross product
normal_vector = np.cross(AB, AC)

# Define the plane equation coefficients
a, b, c = normal_vector
d = -np.dot(normal_vector, A)

# Create a grid to plot the plane
xx, yy = np.meshgrid(np.linspace(-1, 5, 50), np.linspace(-1, 5, 50))
zz = (-a * xx - b * yy - d) / c

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(xx, yy, zz, color='lightblue', alpha=0.5, rstride=100, cstride=100)

# Plot the points A, B, and C
ax.scatter(*A, color='red', s=100, label='A (1,2,3)')
ax.scatter(*B, color='green', s=100, label='B (3,4,5)')
ax.scatter(*C, color='blue', s=100, label='C (2,1,4)')

# Plot the normal vector
ax.quiver(*A, *normal_vector, color='black', length=2, linewidth=2, label='Normal Vector')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set title and legend
ax.set_title('Plane through points A, B, and C')
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the normal vectors
normal_1 = np.array([1, 1, 1])
normal_2 = np.array([1, 0, 0])

# Normalize the vectors to get direction vectors
normal_1_unit = normal_1 / np.linalg.norm(normal_1)
normal_2_unit = normal_2 / np.linalg.norm(normal_2)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the normal vectors
ax.quiver(0, 0, 0, normal_1_unit[0], normal_1_unit[1], normal_1_unit[2], color='r', label='Normal Vector (x + y + z = 1)')
ax.quiver(0, 0, 0, normal_2_unit[0], normal_2_unit[1], normal_2_unit[2], color='b', label='Normal Vector (x = 0)')

# Set the limits of the axes
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])

# Label the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a legend
ax.legend()

# Set the title
ax.set_title('Visualization of the Angle Between Planes')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the normal vector
normal_vector = np.array([1, 1, 1])

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the normal vector
ax.quiver(0, 0, 0, normal_vector[0], normal_vector[1], normal_vector[2], color='r', length=1.5, normalize=True, label='Normal Vector (1, 1, 1)')

# Plot the plane x + y + z = 1
# We need to plot the plane by defining a mesh grid
x = np.linspace(-1, 2, 10)
y = np.linspace(-1, 2, 10)
X, Y = np.meshgrid(x, y)
Z = 1 - X - Y  # Solving for Z from x + y + z = 1

# Plot the plane
ax.plot_surface(X, Y, Z, color='lightblue', alpha=0.5)

# Set the limits of the axes
ax.set_xlim([-1, 2])
ax.set_ylim([-1, 2])
ax.set_zlim([-1, 2])

# Label the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a legend
ax.legend()

# Set the title
ax.set_title('Visualization of the Normal Vector to the Plane x + y + z = 1')

# Show the plot
plt.show()
```
